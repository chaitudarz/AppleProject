```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from fontTools.otlLib.optimize.gpos import Cluster
from sklearn.cluster import DBSCAN, KMeans, HDBSCAN
```

### DBSCAN 
1. The key idea is that for each point of a cluster, the neighbourhood of a given radius has to contain at least a minimum number of points. [Ref](https://www.geeksforgeeks.org/dbscan-clustering-in-ml-density-based-clustering/)

## DBSCAN Parameters
DBSCAN has two parameters. The first is ε, epsilon (“esp”), which defines the maximum distance allowed between two points within the same cluster. The second is minimum samples (“MinPts”), which defines the minimum number of data points required to form a distinct cluster. So, MinPts is the minimum number of neighbors contained within a cluster with radius/max length of esp.

```python
# Data preprocessing
# Converting the text file to CSV for easy accessing the rows and columns

input_txt_file = "/home/chaitu/Desktop/VR_predict/predict6_0/labels/cam_1_000619_000331836957_reprojected_0.txt"
output_csv_file = input_txt_file[:-4] + ".csv"

# Read the YOLOV8 output txt file
# Column names: class, x_center, y_center, width, height

columns = ['Class', 'X_center', 'Y_center', 'Width', 'Height']

df = pd.read_csv(input_txt_file, sep = " ", header= None, names = columns)

#Save the DataFrame (df) as a CSV file
df.to_csv(output_csv_file, index = False)
```

```python
#clustering 

csv_path = "/home/chaitu/Desktop/VR_predict/predict6_1/labels/cam_1_000619_000331836957_reprojected_1.csv"
#csv_path = "/home/chaitu/Desktop/inmach_predict/labels/test1_inmach.csv"

bounding_boxes_df = pd.read_csv(csv_path)

bbox_centers_normalized = bounding_boxes_df[['X_center', 'Y_center']].values

# Convert bounding boxes to image pixel coordinates
image_width = 2048 # 1922 #2048
image_height = 2048 # 1200 #2048
bbox_centers = np.array([(bbox[0] * image_width, bbox[1] * image_height) for bbox in bbox_centers_normalized])

# Additional snippet for Vertical scaling 
# Apply horitzontal bias by scaling the X-coordinate
horitzontal_scale_factor = 5
bbox_centers_scaled = np.copy(bbox_centers)
bbox_centers_scaled[:, 0] = bbox_centers_scaled[:, 0] * horitzontal_scale_factor

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=450, min_samples=4) #450, 3
clusters = dbscan.fit_predict(bbox_centers_scaled)

# Visualize the clustering
plt.figure(figsize=(10,6))
for cluster_id in np.unique(clusters):
    cluster_points = bbox_centers[clusters == cluster_id]
    if cluster_id == -1:
        # Noise points
        plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label = 'Noise', color='red', marker='x')
        
    else:
        plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label =f'Tree {cluster_id +1}')

plt.title('Clustering Apples  by Tree using DBSCAN (with vertical Bias)')
plt.xlabel('X (pixels)')
plt.ylabel('Y (pixels)')
plt.legend()
plt.gca().invert_yaxis() # Invert Y axis for image coordinate system
plt.show()

# Print Cluster results
print(f"Cluters found: {np.unique(clusters)}")
        
```

### K-Nearest Neighbors to identify MinPts, & eps

```python
from sklearn.neighbors import NearestNeighbors

neighbors = NearestNeighbors(n_neighbors=2)
neighbors_fit = neighbors.fit(bbox_centers)
distances, indices = neighbors_fit.kneighbors(bbox_centers)

distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.plot(distances)
```
